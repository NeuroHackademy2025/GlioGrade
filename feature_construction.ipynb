```python
### Create the features chart, using voxel to samples

import os
import numpy as np
import pandas as pd
import nibabel as nib
import scipy.stats
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GroupKFold
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# First, let's create the voxel feature extraction for the full dataset
def extract_brain_voxel_features(patient_id, base_path='../shared/data/PDGM/UCSF-PDGM-v5/UCSF-PDGM-v5/'):
    """Extract features from brain voxels for one patient"""
    
    patient_id_padded = patient_id.split('-')[-1].zfill(4)
    
    # File paths
    fa_path = os.path.join(base_path, f"UCSF-PDGM-{patient_id_padded}_nifti", 
                         f'UCSF-PDGM-{patient_id_padded}_DTI_eddy_FA.nii.gz')
    tumor_path = os.path.join(base_path, f"UCSF-PDGM-{patient_id_padded}_nifti", 
                             f'UCSF-PDGM-{patient_id_padded}_tumor_segmentation.nii.gz')
    brain_path = os.path.join(base_path, f"UCSF-PDGM-{patient_id_padded}_nifti", 
                             f'UCSF-PDGM-{patient_id_padded}_brain_segmentation.nii.gz')
    
    # Check if files exist
    if not all(os.path.exists(p) for p in [fa_path, tumor_path, brain_path]):
        return None, None
    
    try:
        # Load data
        fa_data = nib.load(fa_path).get_fdata()
        tumor_mask = nib.load(tumor_path).get_fdata()
        brain_mask = nib.load(brain_path).get_fdata()
        
        # Get brain-only voxels (excluding tumor)
        brain_indices = np.where(brain_mask > 0.5)
        tumor_binary = tumor_mask > 0
        
        # Filter out tumor voxels
        brain_not_tumor_mask = np.zeros_like(brain_mask, dtype=bool)
        for i in range(len(brain_indices[0])):
            x, y, z = brain_indices[0][i], brain_indices[1][i], brain_indices[2][i]
            if not tumor_binary[x, y, z]:
                brain_not_tumor_mask[x, y, z] = True
        
        brain_only_indices = np.where(brain_not_tumor_mask)
        n_brain_voxels = len(brain_only_indices[0])
        
        if n_brain_voxels == 0:
            return None, None
        
        # Sample voxels if too many
        max_voxels = 5000
        if n_brain_voxels > max_voxels:
            np.random.seed(42)  # For reproducibility
            sample_idx = np.random.choice(n_brain_voxels, max_voxels, replace=False)
            brain_only_indices = (brain_only_indices[0][sample_idx], 
                                 brain_only_indices[1][sample_idx], 
                                 brain_only_indices[2][sample_idx])
            n_brain_voxels = max_voxels
        
        # Compute tumor center
        tumor_coords = np.where(tumor_mask > 0)
        if len(tumor_coords[0]) == 0:
            return None, None
            
        tumor_center = np.array([np.mean(tumor_coords[0]), 
                                np.mean(tumor_coords[1]), 
                                np.mean(tumor_coords[2])])
        
        # Extract features
        features = []
        for i in range(n_brain_voxels):
            x, y, z = brain_only_indices[0][i], brain_only_indices[1][i], brain_only_indices[2][i]
            
            # Core features
            fa_val = fa_data[x, y, z]
            norm_x = x / fa_data.shape[0]
            norm_y = y / fa_data.shape[1]
            norm_z = z / fa_data.shape[2]
            dist_to_tumor = np.sqrt(np.sum((np.array([x,y,z]) - tumor_center)**2))
            
            # Local neighborhood
            x_min, x_max = max(0, x-1), min(fa_data.shape[0], x+2)
            y_min, y_max = max(0, y-1), min(fa_data.shape[1], y+2)
            z_min, z_max = max(0, z-1), min(fa_data.shape[2], z+2)
            
            neighborhood = fa_data[x_min:x_max, y_min:y_max, z_min:z_max]
            local_mean = np.mean(neighborhood)
            local_std = np.std(neighborhood)
            local_max = np.max(neighborhood)
            local_min = np.min(neighborhood)
            
            # Gradients
            grad_x = (fa_data[min(x+1, fa_data.shape[0]-1), y, z] - 
                     fa_data[max(x-1, 0), y, z]) / 2
            grad_y = (fa_data[x, min(y+1, fa_data.shape[1]-1), z] - 
                     fa_data[x, max(y-1, 0), z]) / 2
            grad_z = (fa_data[x, y, min(z+1, fa_data.shape[2]-1)] - 
                     fa_data[x, y, max(z-1, 0)]) / 2
            grad_mag = np.sqrt(grad_x**2 + grad_y**2 + grad_z**2)
            
            features.append([fa_val, norm_x, norm_y, norm_z, dist_to_tumor,
                           local_mean, local_std, local_max, local_min, grad_mag])
        
        return np.array(features), n_brain_voxels
        
    except Exception as e:
        print(f"Error processing {patient_id}: {e}")
        return None, None

# Load metadata
print("Loading metadata...")
df = pd.read_csv('PDGM/UCSF-PDGM-metadata_v5.csv')
print(f"Total patients in metadata: {len(df)}")
print(f"WHO grade distribution:\n{df['WHO CNS Grade'].value_counts().sort_index()}")

# Check all clinical and molecular variables
print("\nClinical and molecular variables summary:")
print(f"Age at MRI: {df['Age at MRI'].describe()}")
print(f"Sex distribution:\n{df['Sex'].value_counts()}")
print(f"Overall Survival (OS): {df['OS'].describe()}")
print(f"\nMGMT status distribution:\n{df['MGMT status'].value_counts()}")
print(f"MGMT index: {df['MGMT index'].describe()}")
print(f"\n1p/19q status distribution:\n{df['1p/19q'].value_counts()}")
print(f"\nEOR (Extent of Resection) distribution:\n{df['EOR'].value_counts()}")

# Process all patients
print("\nProcessing all patients...")
all_features = []
all_labels = []
all_patient_ids = []
all_ages = []
all_sexes = []
all_os = []
all_mgmt_status = []
all_mgmt_index = []
all_1p19q = []
all_eor = []

feature_names = ['fa_value', 'norm_x', 'norm_y', 'norm_z', 'dist_to_tumor',
                'local_mean', 'local_std', 'local_max', 'local_min', 'gradient_magnitude']

# Process in batches to save memory
processed_count = 0
failed_count = 0

for idx, row in tqdm(df.iterrows(), total=len(df), desc="Processing patients"):
    patient_id = row['ID']
    who_grade = row['WHO CNS Grade']
    age = row['Age at MRI']
    sex = row['Sex']
    os_days = row['OS']
    mgmt_status = row['MGMT status']
    mgmt_index = row['MGMT index']
    one_p_19q = row['1p/19q']
    eor = row['EOR']
    
    if pd.isna(who_grade):
        continue
        
    features, n_voxels = extract_brain_voxel_features(patient_id)
    
    if features is not None:
        all_features.extend(features)
        all_labels.extend([int(who_grade)] * len(features))
        all_patient_ids.extend([patient_id] * len(features))
        all_ages.extend([age] * len(features))
        all_sexes.extend([sex] * len(features))
        all_os.extend([os_days] * len(features))
        all_mgmt_status.extend([mgmt_status] * len(features))
        all_mgmt_index.extend([mgmt_index] * len(features))
        all_1p19q.extend([one_p_19q] * len(features))
        all_eor.extend([eor] * len(features))
        processed_count += 1
    else:
        failed_count += 1
    
    # Save periodically to avoid memory issues
    if (idx + 1) % 50 == 0:
        print(f"\nProcessed {processed_count} patients successfully, {failed_count} failed")

# Convert to arrays
X = np.array(all_features)
y = np.array(all_labels)

print(f"\nFinal dataset shape: {X.shape}")
print(f"Successfully processed: {processed_count} patients")
print(f"Failed: {failed_count} patients")
print(f"Total voxels: {len(X)}")
print(f"Grade distribution in voxels: {np.unique(y, return_counts=True)}")

# Create and save the comprehensive CSV
print("\nCreating comprehensive CSV file...")
voxel_df = pd.DataFrame(X, columns=feature_names)
voxel_df['WHO_grade'] = y
voxel_df['patient_id'] = all_patient_ids
voxel_df['age_at_mri'] = all_ages
voxel_df['sex'] = all_sexes
voxel_df['overall_survival_days'] = all_os
voxel_df['mgmt_status'] = all_mgmt_status
voxel_df['mgmt_index'] = all_mgmt_index
voxel_df['1p19q'] = all_1p19q
voxel_df['eor'] = all_eor

# Save to CSV
csv_filename = 'brain_voxel_features_full_clinical.csv'
voxel_df.to_csv(csv_filename, index=False)
print(f"Saved voxel features with all clinical data to {csv_filename}")
print(f"File size: {os.path.getsize(csv_filename) / 1024 / 1024:.1f} MB")

# Display sample
print("\nSample of the data:")
print(voxel_df.head())
print(f"\nPatients in dataset: {voxel_df['patient_id'].nunique()}")

# Create patient-level summary
print("\nCreating patient-level summary...")
patient_summary = voxel_df.groupby('patient_id').agg({
    'WHO_grade': 'first',
    'age_at_mri': 'first',
    'sex': 'first',
    'overall_survival_days': 'first',
    'mgmt_status': 'first',
    'mgmt_index': 'first',
    '1p19q': 'first',
    'eor': 'first',
    'fa_value': ['mean', 'std', 'min', 'max'],
    'dist_to_tumor': ['mean', 'std', 'min'],
    'gradient_magnitude': ['mean', 'std']
}).reset_index()

# Flatten column names
patient_summary.columns = ['_'.join(col).strip('_') if col[1] else col[0] 
                          for col in patient_summary.columns.values]

patient_summary_filename = 'patient_level_summary_full.csv'
patient_summary.to_csv(patient_summary_filename, index=False)
print(f"Saved patient-level summary to {patient_summary_filename}")

# Analyze molecular markers by grade
print("\nMolecular markers by WHO grade:")
unique_patients = voxel_df.drop_duplicates('patient_id')

# MGMT status by grade
mgmt_by_grade = pd.crosstab(unique_patients['WHO_grade'], unique_patients['mgmt_status'], normalize='index')
print("\nMGMT status by grade (proportions):")
print(mgmt_by_grade)

# 1p/19q by grade
onep19q_by_grade = pd.crosstab(unique_patients['WHO_grade'], unique_patients['1p19q'], normalize='index')
print("\n1p/19q status by grade (proportions):")
print(onep19q_by_grade)

# EOR by grade
eor_by_grade = pd.crosstab(unique_patients['WHO_grade'], unique_patients['eor'], normalize='index')
print("\nExtent of resection by grade (proportions):")
print(eor_by_grade)

# Visualize molecular markers
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# MGMT status by grade
mgmt_counts = pd.crosstab(unique_patients['WHO_grade'], unique_patients['mgmt_status'])
mgmt_counts.plot(kind='bar', ax=axes[0,0], rot=0)
axes[0,0].set_xlabel('WHO Grade')
axes[0,0].set_ylabel('Count')
axes[0,0].set_title('MGMT Status by WHO Grade')
axes[0,0].legend(title='MGMT Status')

# MGMT index by grade
for grade in [2, 3, 4]:
    grade_data = unique_patients[unique_patients['WHO_grade'] == grade]['mgmt_index'].dropna()
    axes[0,1].hist(grade_data, alpha=0.5, label=f'Grade {grade}', bins=20, density=True)
axes[0,1].set_xlabel('MGMT Index')
axes[0,1].set_ylabel('Density')
axes[0,1].set_title('MGMT Index Distribution by WHO Grade')
axes[0,1].legend()

# 1p/19q by grade
onep19q_counts = pd.crosstab(unique_patients['WHO_grade'], unique_patients['1p19q'])
onep19q_counts.plot(kind='bar', ax=axes[1,0], rot=0)
axes[1,0].set_xlabel('WHO Grade')
axes[1,0].set_ylabel('Count')
axes[1,0].set_title('1p/19q Status by WHO Grade')
axes[1,0].legend(title='1p/19q Status')

# EOR by grade
eor_counts = pd.crosstab(unique_patients['WHO_grade'], unique_patients['eor'])
eor_counts.plot(kind='bar', ax=axes[1,1], rot=0)
axes[1,1].set_xlabel('WHO Grade')
axes[1,1].set_ylabel('Count')
axes[1,1].set_title('Extent of Resection by WHO Grade')
axes[1,1].legend(title='EOR', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig('molecular_markers_by_grade.png')
plt.show()

# Survival analysis by molecular markers
plt.figure(figsize=(12, 5))

# Survival by MGMT status
plt.subplot(1, 2, 1)
mgmt_groups = unique_patients.groupby('mgmt_status')['overall_survival_days']
for name, group in mgmt_groups:
    if name in ['positive', 'negative']:  # Focus on clear cases
        plt.hist(group.dropna(), alpha=0.5, label=f'MGMT {name}', bins=30, density=True)
plt.xlabel('Overall Survival (days)')
plt.ylabel('Density')
plt.title('Survival by MGMT Status')
plt.legend()
plt.xlim(0, 3000)

# Survival by 1p/19q status
plt.subplot(1, 2, 2)
onep19q_groups = unique_patients.groupby('1p19q')['overall_survival_days']
for name, group in onep19q_groups:
    if pd.notna(name) and name != 'unknown':
        plt.hist(group.dropna(), alpha=0.5, label=f'1p/19q: {name}', bins=30, density=True)
plt.xlabel('Overall Survival (days)')
plt.ylabel('Density')
plt.title('Survival by 1p/19q Status')
plt.legend()
plt.xlim(0, 3000)

plt.tight_layout()
plt.savefig('survival_by_molecular_markers.png')
plt.show()
```
